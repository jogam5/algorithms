## Section 1

**(a)** Note that **$n\,log(n^3) = 3\,n\,log(n)$**. If $c = 5$ and $n_0 = 1$, we have $3\,n\,log(n) \leq 5\,n\,log(n)$ for all $n \geq 1$, therefore $f(n) = O(g(n))$. To prove big Omega, pick any c below or equal to 3, for instance if $c = 2$ and $n_0=1$, we have $3\,n\,log(n) \geq 2\,n\,log(n)$ for all $n\geq1$, therefore $f(n) = \Omega(g(n))$. As a result, we conclude that $f(n)=\Theta(g(n))$.
